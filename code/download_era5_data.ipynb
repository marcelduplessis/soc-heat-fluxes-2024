#### Download ERA5 data for the State of the Climate Report

```python
from functions.download_era5_cdsapi import download_era5_cdsapi

import numpy as np
import calendar
```

```python
dwn_path = '/Volumes/LaCie/Work/data/era5/2024_soc_report/'

years = np.arange(1998, 2003, 1)
months = np.arange(1, 13, 1)

var = [
    # 'surface_latent_heat_flux',
    # 'surface_sensible_heat_flux',    
    # '2m_dewpoint_temperature',
    # '2m_temperature',
    'sea_ice_cover',
    # '10m_v_component_of_wind',
    # '10m_u_component_of_wind',
    # 'sea_surface_temperature',
    # 'skin_temperature',
    # 'surface_pressure',
    # 'surface_net_solar_radiation',
    # 'surface_net_thermal_radiation',
    # 'total_precipitation',
    # 'evaporation'
      ]
```

```python
# download hourly files for 2023

for v in var:
    
    for year in years:
        
        for month in months:
            
            # Determine the number of days in the current month
            num_days = calendar.monthrange(year, month)[1]
            
            # Create a list of days using list comprehension
            days = [f'{d:02}' for d in range(1, num_days + 1)]
            
            # Format month as a string with leading zero if needed
            month_str = f'{month:02}'
            
            # Download Monthly Data
            # dwn_era5.download_era5_cdsapi(var=v, days=days, month=month_str, year=str(year), dataset='reanalysis-era5-single-levels-monthly-means', product_type='monthly_averaged_reanalysis', path=dwn_path)
            download_era5_cdsapi(var=v, days=days, month=month_str, year=str(year), dataset='reanalysis-era5-single-levels', product_type='reanalysis', path=dwn_path)
```
